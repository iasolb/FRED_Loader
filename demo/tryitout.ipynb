```python
import sys, os, time
sys.path.insert(0, os.path.abspath(".."))
from load import pull_fred
from utils import Config
    
from series import (
    # Top-level categories
    INFLATION, LABOR, ALL_SERIES,
    # Subcategories directly
    CPI, TREASURIES, ENERGY, JOLTS,
    # Or via the lookup
    SUBCATEGORIES,
)
```

```python
configs = []
start_date_override = "2015-01-01"
# ── By subcategory import ─────────────────────────────────
config1 = Config(
    filename="cpi_and_treasuries.csv",
    output_path="example_outputs",
    start=start_date_override,
    series={**CPI, **TREASURIES},
)
configs.append(config1)
# ── By category import ────────────────────────────────────
config2 = Config(
    filename="inflation_and_labor.csv",
    output_path="example_outputs",
    start=start_date_override,
    series={**INFLATION, **LABOR},
)
configs.append(config2)
# ── Via SUBCATEGORIES lookup ──────────────────────────────
config3 = Config(
    filename="energy_and_jolts.csv",
    output_path="example_outputs",
    start=start_date_override,
    series={
        **SUBCATEGORIES["COMMODITIES"]["ENERGY"],
        **SUBCATEGORIES["LABOR"]["JOLTS"],
    },
)
configs.append(config3)
# ── Mix subcategories with one-offs ───────────────────────
config4 = Config(
    filename="custom.csv",
    output_path="example_outputs",
    start=start_date_override,
    series={
        **CPI,
        **ENERGY,
        "UNRATE": ("Unemployment_Rate", "M"),
    },
)
configs.append(config4)
```

```python
# Subset Checks 
for config in configs:
    print(f"Pulling data for {config.FILENAME}...")
    pull_fred(config)
    print(f"Finished pulling data for {config.FILENAME}")
    time.sleep(3)

print("All done!")
print("Check the 'demo/example_outputs' directory for the downloaded CSV files.")
```

```python
# ── Everything ────────────────────────────────────────────
configfull = Config(
    filename="all.csv",
    output_path="./data",
    series={**ALL_SERIES},
)
configs.append(configfull)
print(f"Pulling data for {configfull.FILENAME}...")
pull_fred(configfull)
print(f"Finished pulling data for {configfull.FILENAME}")
```

```python
for config in configs:
    file_path = os.path.join(config.OUTPUT_PATH, config.FILENAME)
    if os.path.exists(file_path):
        size_mb = os.path.getsize(file_path) / (1024 * 1024)
        print(f"{config.FILENAME}: {size_mb:.2f} MB")
    else:
        print(f"{config.FILENAME}: File not found.")
```
